```python
# Importing necessary libraries for analysis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Getting dataset for Men's results
mens_results = pd.read_csv(r'D:\Data Journey\Arewa-Data-Science-Fellowship\Arewa-DS-Projects\Data Sets\results.csv')
print(mens_results.head(5))
print('Number of rows and columns in dataset is', mens_results.shape)
print('List of columns of dataset is', mens_results.columns)
```

```python
# Exploratory Data Analysis
print('Dataset Information',mens_results.info())

print(mens_results.describe())

# Misisng data
print(mens_results.isna().sum())
```

```python
# Understanding tournament column for mens results
# Filter for actual FIFA World Cup matches (exclude qualifiers)
print(mens_results['tournament'].nunique())
filtered_men = mens_results[(mens_results['tournament'].str.contains('FIFA World Cup', case=False)) & (~mens_results['tournament'].str.contains('qualification', case=False))].copy()

# Ensure 'date' is datetime
filtered_men ['date'] = pd.to_datetime(filtered_men ['date'])

# Keep matches from 2002 onwards
filtered_men  = filtered_men [filtered_men['date'] >= '2002-01-01']

# Total score
filtered_men ['total_score'] = filtered_men ['home_score'] + filtered_men ['away_score']

# Check results
filtered_men 
```
