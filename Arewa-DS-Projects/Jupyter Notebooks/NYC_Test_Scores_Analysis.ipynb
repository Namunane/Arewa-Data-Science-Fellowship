```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

schools_results = pd.read_csv(r'D:\Data Journey\Data Science\Data Camp\Arewa-DS-Projects\Data Sets\schools.csv')
schools_results.head(5)
```

```python
# Finding schools with the top maths results
max_result = 800
threshold = 0.8 * max_result
best_maths_schools = (schools_results[schools_results['average_math'] >= threshold][['school_name', 'average_math']].sort_values('average_math',ascending= False).reset_index(drop=False))

# Display the results
print("Top performing schools in mathematics (scores >= 640):")
print(best_maths_schools)
```

```python
# Finding Top 10 schools based on combined SAT scores
# Use subsetting; 2 square brackets to apply statistical calculations
schools_results['total_SAT'] = schools_results[['average_math', 'average_reading', 'average_writing']].sum(axis=1)
top_schools = schools_results[['school_name', 'total_SAT']].sort_values('total_SAT',ascending= False).reset_index(drop=False)
top_schools.head(10)
```

```python
# Calculating borough std
borough_std = schools_results.groupby('borough')['total_SAT'].std()
borough_std_max = borough_std.idxmax()
borough_std_max_value = borough_std.max()

print("Standard deviation of total_SAT by borough:")
print(borough_std)

# f-string print
print(f"\nBorough with the largest standard deviation: {borough_std_max} ({borough_std_max_value:.2f})")
```

```python
# Find the borough with the largest std of total_SAT
borough_std = schools_results.groupby('borough')['total_SAT'].std()
borough_with_max_std = borough_std.idxmax()

# Calculate required statistics for that borough
borough_data = schools_results[schools_results['borough'] == borough_with_max_std]
num_schools = borough_data.shape[0]
average_SAT = borough_data['total_SAT'].mean()
std_SAT = borough_data['total_SAT'].std()

# Create the DataFrame
largest_std_dev= pd.DataFrame({
    'borough': [borough_with_max_std],
    'num_schools': [num_schools],
    'average_SAT': [average_SAT],
    'std_SAT': [std_SAT]
})

largest_std_dev
```
