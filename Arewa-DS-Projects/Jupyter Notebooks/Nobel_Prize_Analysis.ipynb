```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r'D:\Data Journey\Data Science\Data Camp\Arewa-DS-Projects\Data Sets\nobel.csv')
# Accessing the data set and setting the index as the first column

data.head(5)
```

```python
# Understanding the data columns
data.describe()
data.info()
data.shape # Called to find the rows and columns, dont include the brackets
```

```python
# Finding the most common gender and birth country

# Most common gender
gender_counts = data['sex'].value_counts()
most_common_gender = gender_counts.idxmax()
print("Gender Counts:\n", gender_counts)
print("Most Common Gender:", most_common_gender)

# Most common birth country
country_counts = data['birth_country'].value_counts()
most_common_country = country_counts.idxmax()
print("Country Counts:\n", country_counts)
print("Most Common Country:", most_common_country)
```

```python
# Decade with the highest ratio of US born Nobel Prize winners in all categories

# Creating a column for US Nobel prize winners
data['US Winners'] = data['birth_country'] == 'United States of America'
data.head(3)

```

```python
# Creating a decade column
# Understanding the year column
data['year'].info()
data['year'].describe() 

data['Decade'] = (data['year']// 10) * 10

data.head(10)
```

```python
print(data.columns)
```

```python
# Calculate the average (mean) of US-born Nobel Prize winners by decade
us_winners_by_decade = data.groupby('Decade')['US Winners'].mean()

# Find the decade with the highest ratio
max_decade = us_winners_by_decade.idxmax()
max_ratio = us_winners_by_decade.max()

# Print results in a clean format
print("Average ratio of US-born Nobel Prize winners by decade:")
print(us_winners_by_decade)

print(f"The decade with the highest ratio of US-born winners is: {max_decade} with a ratio of {max_ratio}")
```

```python
# Creating a relational plot to visualize the trend of average ratios of US-born Nobel Prize winners
sns.lineplot(x=us_winners_by_decade.index, y=us_winners_by_decade.values)
plt.xlabel('Decade')
plt.ylabel('Ratio of US-born Nobel Prize winners')
plt.title('A graph showing the trend of US-born Nobel Prize winners')
             
plt.show()
```

```python
# Convert the 'sex' column to a numerical 'Gender' column (1 for Female, 0 for Male)
data['Gender'] = data['sex'].apply(lambda y: 1 if y == 'Female' else 0)

# Group by decade to calculate the average ratio of female laureates per decade
female_ratio_by_decade = data.groupby('Decade')['Gender'].mean()

# Find the decade with the highest ratio of female winners
max_female_decade = female_ratio_by_decade.idxmax()
max_female_ratio = female_ratio_by_decade.max()

# Print results in a clean format
print("📊 Average ratio of female Nobel Prize winners by decade:")
print(female_ratio_by_decade)

print(f"\n The decade with the highest ratio of female winners is: {max_female_decade} "
      f"with a ratio of {max_female_ratio}")
```

```python
# Filter for female winners
female_winners = data[data['sex'] == 'Female']

# Sort by year to find the first
first_female_winner = female_winners.sort_values('year').head(1)

print(first_female_winner[['full_name','birth_country','year']])
```
