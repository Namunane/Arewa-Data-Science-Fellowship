```python
## Uncovering interests facts about using line, scatter and bar plots

# Importing necessary libraries for analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'D:\Data Journey\Data Science\Arewa Data Science Fellowship\Assignments\Datasets\birds.csv')
data.head()
```

```python
# Investigating Snow Geese

# Counting occurrences using value_counts on the 'Name' column
snow_goose_count = data['Name'].value_counts().get('Snow goose', 0)
print(f"Snow goose appears {snow_goose_count} times in the dataset.")

```

```python
# Getting information on Snow Goose
Snow_goose_details = data[data['Name'] == 'Snow goose']
print(Snow_goose_details)

# Getting data on numericals
print(Snow_goose_details[['']])
```

```python
# Getting columns
columns = data[data['Name'] == 'Snow goose'].columns
columns
```

```python
# Visualization of snow goose characteristics in comparison to other birds
Unique_birds = data['Name'].unique()
Number_of_unique_birds = len(Unique_birds)
print(Number_of_unique_birds)


plt.figure(figsize=(8,5))
x_axis = data['Name']
y_axis = data['Name'].value_counts()
plt.bar(x_axis, y_axis)
plt.show()
```

```python
Unique_birds = data['Name'].unique()
Number_of_unique_birds = len(Unique_birds)
print(Number_of_unique_birds)
```

```python
# Count occurrences of each bird name
bird_counts = data['Name'].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(bird_counts.index, bird_counts.values, color=['red' if name == 'Snow goose' else 'skyblue' for name in bird_counts.index])
plt.xlabel('Bird Name')
plt.ylabel('Count')
plt.title('Bird Species Counts (Highlighting Snow Goose)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Compare Snow Goose to other birds (if weight and wingspan available)

sns.scatterplot(data=data, x='MaxWingspan', y='MaxBodyMass', hue=(data['Name'] == 'snow goose'))
plt.title("Weight vs WingSpan (Highlighting Snow Goose)")
```
