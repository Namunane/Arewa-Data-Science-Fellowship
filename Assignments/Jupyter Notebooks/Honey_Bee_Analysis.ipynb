```python
## Uncovering interests facts about using line, scatter and bar plots

# Importing necessary libraries for analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'D:\Data Journey\Data Science\Arewa Data Science Fellowship\Assignments\Datasets\honey.csv')
data
```

```python
# Missing Values & Data Cleaning
# Check for missing values
print(data.isnull().sum())
# Optionally drop or fill missing values
# data = data.dropna()
```

```python
# Aggregate colony numbers by year
colony_trend = data.groupby('state')['priceperlb'].sum()
plt.figure(figsize=(10,5))
plt.plot(colony_trend.index, colony_trend.values, marker='o')
plt.title('Price Variation per State')
plt.xlabel('State')
plt.ylabel('Price')
plt.show()
```

```python
print(data['state'].unique())
```

```python
# Top states by honey production
top_states = data.groupby('state')['totalprod'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_states.index, y=top_states.values)
plt.title('Top 10 States by Total Honey Production')
plt.xlabel('State')
plt.ylabel('Honey Production')
plt.show()
```

```python
# Production trends by State
ND_data = data[data['state'] == 'ND']
plt.figure(figsize=(10,5))
plt.plot(ND_data['year'], ND_data['prodvalue'], marker='o', color = 'red')
plt.title('ND production value trends')
plt.xlabel('Year')
plt.ylabel('Production Value')
plt.show()
```
