```python
# General Time Series Analysis Template
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

```

```python
# Reading dataset
df = pd.read_csv(r"D:\Data Journey\Arewa-Data-Science-Fellowship\Assignments\Datasets\AirPassengers.csv")

# Renaming dataset columns
df.columns = ['Date', 'Number of Passengers']
# Converting date data type to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# --- Step 1: Plot raw series ---
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Number of Passengers'], marker='', color='b') 
plt.title("Number of US Airline passengers from 1949 to 1960")
plt.xlabel("Time")
plt.ylabel("Number of Passengers") 
plt.grid(True)
plt.show()


```

```python
# --- Step 2: Summary statistics ---
print("Summary:")
print(df['Number of Passengers'].describe())  # Corrected column name

# --- Step 3: Seasonal decomposition ---
# period depends on dataset (e.g., 12 for monthly data with yearly seasonality)
decomp = seasonal_decompose(df['Number of Passengers'], model='multiplicative', period=12)  # Corrected column name

fig = decomp.plot()
fig.set_size_inches(12, 8)
plt.show()

# --- Step 4: Observations ---
# Trend, seasonality, residual can be accessed via:
trend = decomp.trend
seasonality = decomp.seasonal
residuals = decomp.resid
```
